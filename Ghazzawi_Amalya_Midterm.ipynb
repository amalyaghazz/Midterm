```python
import pandas as pd

df = pd.read_csv('amazon-orders.csv')
df.head()
```

```python
df.shape
```

```python
df = df.fillna(0)
df.head()
```

```python
df["Total Charged"] = df["Total Charged"].str.replace('$','').astype(float)
df.head()
```

```python
df["Total Charged"].sum()
```

```python
df["Total Charged"].mean()
```

```python
df["Total Charged"].median()
```

```python
df["Total Charged"].min()
```

```python
df["Total Charged"].max()
```

```python
df["Tax Charged"] = df["Tax Charged"].str.replace('$','').astype(float)
df.head()
```

```python
df["Tax Charged"].sum()
```

```python
df["Tax Charged"].sum() / df["Total Charged"].sum()
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df.head()
```

```python
df.plot.bar(x='Order Date', y='Total Charged', rot=90, figsize=(20,10), color='c')
```

```python
daily_orders = df.groupby('Order Date').sum()["Total Charged"]
daily_orders.head()
```

```python
daily_orders.plot.bar(figsize=(20,10), color='c')
```

```python
df['Order Date'] = pd.to_datetime(df['Order Date'])

df["Year"] = pd.DatetimeIndex(df['Order Date']).year
df['Month Number'] = pd.DatetimeIndex(df['Order Date']).month

month = { 
            1:'January', 
            2:'February', 
            3:'March', 
            4:'April', 
            5:'May', 
            6:'June', 
            7:'July', 
            8:'August', 
            9:'September', 
            10:'October', 
            11:'November', 
            12:'December'
        }

df["Month"] = df['Order Date'].dt.month.map(month)
df.head()
```

```python
monthly_orders = df.groupby('Month').sum()["Total Charged"]
monthly_orders
```

```python
monthly_orders.plot.bar(figsize=(20,10), color='m')
```

```python
item_category = ["Beauty", "Health", "Home", "Clothing", "Beauty", "Beauty", "Health", "Clothing", "Electronics", "Electronics", "Electronics", "Electronics", "Home", "Home", "Home", "Home", "Health", "Health", "Health", "Health","Clothing","Home","Electronics","Electronics", "Beauty", "Home","Home","Home", "Health","Health","Health","Health","Electronics","Home", "Electronics","Beauty","Home","Beauty","Electronics","Health","Beauty","Beauty","Beauty","Electronics","Home","Home","Home","Home","Health","Beauty","Home","Electronics","Home","Electronics","Clothing","Beauty","Home","Electronics","Health"]
df['Category'] = pd.array(item_category)
df.head()
```

```python
category_sum = df.groupby("Category").sum()
category_sum
final = category_sum.loc[:, ["Tax Charged", "Total Charged"]]
```

```python
final.plot.bar(figsize=(20,10), color='r')
```

```python
charge_final= category_sum.loc[:, ["Total Charged"]]
ax = charge_final.plot.pie(y = 'Total Charged', radius=1.0, autopct='%1.1f%%', title="Amazon Summary by Category")
ax.legend(loc="upper left", bbox_to_anchor=(1.2, 1))
ax.set_ylabel("")

```

```python
tax_final = category_sum.loc[:, ["Tax Charged"]]

bx = tax_final.plot.pie(y = 'Tax Charged', radius=1.0, autopct='%1.1f%%', title="Amazon Summary by Category")
bx.legend(loc="upper left", bbox_to_anchor=(1.2, 1))
bx.set_ylabel("")
```
